# preprocessing=>filling missing value

```python
df['BasePay'].fillna(df['BasePay'].median(), inplace=True)
df['Benefits'].fillna(df['Benefits'].median(), inplace=True)
df['OvertimePay'].fillna(df['OvertimePay'].median(), inplace=True)
df['OtherPay'].fillna(df['OtherPay'].median(), inplace=True)
df.isnull().sum()
```

```python
df['Status'].fillna(df['Status'].fillna(0), inplace=True)
df['Notes'].fillna(df['Notes'].fillna(0), inplace=True)
df.isnull().sum()
```

# visualization after preprocesing

```python
df.hist(bins=50 , figsize=(15,11))
plt.show()
```

```python
import seaborn as sn
plt.figure(figsize=(20,30))
for i, col in enumerate(df.columns):
    if df[col].dtype != 'object':
        ax = plt.subplot(9, 2, i+1)
        sn.kdeplot(df[col], ax=ax)
        plt.xlabel(col)
        
plt.show()
```
